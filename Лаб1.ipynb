```
x = np.linspace(0, 10, 50)
y1 = x
y2 = [i**2 for i in x]
plt.title("y1 = x, y2 = x^2")
plt.xlabel("x")
plt.ylabel("y1, y2")
plt.grid()
plt.plot(x,y1,x,y2)
```

```
x = np.linspace(0, 10, 50)
y1 = x
y2 = [i**2 for i in x]

plt.figure(figsize=(9,9))

plt.subplot(2,1,1)
plt.plot(x, y1)
plt.title("y1 = x, y2 = x^2")
plt.ylabel("y1", fontsize=14)
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(x, y2)
plt.ylabel("y1", fontsize=14)
plt.xlabel("x", fontsize=14)
plt.grid(True)
```

```
n = np.linspace(1, 2, 5)

y = n
y1 = np.log10(n)
y2 = np.sqrt(n)
y3 = n*np.log10(n)
y4 = n**2
y5 = n**3
y6 = 2**n

plt.plot(n, y, label="n")
plt.plot(n, y1, label="lg(n)")
plt.plot(n, y2, label="sqrt(n)")
plt.plot(n, y3, label="n")
plt.plot(n, y4, label="n^2")
plt.plot(n, y5, label="n^3")
plt.plot(n, y6, label="2^n")
plt.ylabel("t", fontsize=14)
plt.xlabel("n", fontsize=14)
plt.grid(True)
plt.legend()
```
